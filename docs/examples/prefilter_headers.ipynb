## Prefilter headers

```python
import pdbufr
```

First, we use the default options (``prefilter_headers=False``) and measure the execution time.

```python
%%time
df = pdbufr.read_bufr("aircraft.bufr",          
                      columns=["aircraftFlightNumber"], filters={"ident": "EU4333"})
len(df)
```

Next, we use ``prefilter_headers=True``. We can see that the data extraction became about 20% faster.

```python
%%time
df = pdbufr.read_bufr("aircraft.bufr",          
                      columns=["aircraftFlightNumber"], filters={"ident": "EU4333"},
                     prefilter_headers=True)
len(df)
```
